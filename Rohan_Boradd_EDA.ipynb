```python
# Basic EDA and Visualization
# Distribution of customer age
if 'Region' in customers.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(y = customers['Region'], order = customers['Region'].value_counts().index, color = 'blue')
    plt.title('Distribution of Customers Region')
    plt.xlabel('Frequency')
    plt.ylabel('Region')
    plt.show()
else:
    print("Column 'Region' not Found in customers dataset. Skipping visualization.")
```

### First insight of Exploratory Data Analysis

```python
# Distribution of product prices
plt.figure(figsize =(8, 6))
sns.histplot(products['Price'], kde= True, bins = 20, color = 'green')
plt.title('Distribution of Products price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analyze Total trasactions amounts
if 'Quantity' in transactions.columns and 'Price' in transactions.columns:
    transactions['TotalAmount'] = transactions['Quantity'] * transactions['Price']
    plt.figure(figsize =(8, 6))
    sns.histplot(transactions['TotalAmount'], kde = True, bins = 20, color= 'purple')
    plt.title('Distribution of Total Transactions Amounts')
    plt.xlabel('Transaction Amount')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("Columns 'Quantity' or 'Price' not found in Transactions dataset.")
```

### Seconds insight

```python
top_products = transactions.groupby('ProductID')['Quantity'].sum().nlargest(10)
top_product_names = products[products['ProductID'].isin(top_products.index)]
plt.figure(figsize=(10, 5))
sns.barplot(x=top_products.index, y=top_products.values, palette="rocket")
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Product ID")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
transactions_with_regions = transactions.merge(customers[['CustomerID', 'Region']], on='CustomerID', how='left')
revenue_by_region = transactions_with_regions.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
sns.barplot(x=revenue_by_region.index, y=revenue_by_region.values, palette="mako")
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show()
```

### Insight 
The maximum Revenue is from South America more that 200k USD, Second most is Europe 16k USD,North America and Asia has 15k USD Revenue according to the dataset. We should increase the sales in North America and Asia 

```python
# Additional Analysis: Revenue Contribution of Top 10 Products
top_products_revenue = transactions[transactions['ProductID'].isin(top_products.index)].groupby('ProductID')['TotalValue'].sum()
top_products_revenue_details = top_product_names.merge(top_products_revenue, on='ProductID')

plt.figure(figsize=(10, 5))
sns.barplot(x=top_products_revenue_details['ProductName'], y=top_products_revenue_details['TotalValue'], palette="crest")
plt.title("Revenue Contribution of Top 10 Best-Selling Products")
plt.xlabel("Product Name")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45, ha='right')
plt.show()
```

From the above bargraph we can see that Techpro Headphone are the Best revenue contribution around 19k usd, and second most is TechPro CookBook 18k usd and third most revenue generating is Activewear Rug 18k 

```python
# Additional Analysis: Most Revenue-Generating Products by Region
revenue_by_region_product = transactions_with_regions.groupby(['Region', 'ProductID'])['TotalValue'].sum().reset_index()
most_revenue_products_region = revenue_by_region_product.loc[revenue_by_region_product.groupby('Region')['TotalValue'].idxmax()]
most_revenue_products_region = most_revenue_products_region.merge(products, on='ProductID')

plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='TotalValue', hue='ProductName', data=most_revenue_products_region, palette="tab10")
plt.title("Most Revenue-Generating Products by Region")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.legend(title="Product Name", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Insight 
The most revenue generating products in the speific region are TechPro Headphones with total revenue of 9500 USD, ComfortLiving Sweater and Active wear jeans both are generating the revenue of 6k USD and 6100 USD, Activewear Smartwatch generates 6200 USD.
So this insight of tells us that these four product are the key to the business in the regions we should keep restocking those four products.
